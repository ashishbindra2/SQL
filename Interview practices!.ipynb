#### -- Create the Student table
    `CREATE TABLE IF NOT EXISTS Student (
        STUDENT_ID INT PRIMARY KEY,
        FIRST_NAME VARCHAR(50),
        LAST_NAME VARCHAR(50),
        GPA DECIMAL(4,2),
        ENROLLMENT_DATE DATETIME,
        MAJOR VARCHAR(50)
