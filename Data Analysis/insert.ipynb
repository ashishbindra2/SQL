```python
%%sql 
CREATE TABLE Instructor (
    ins_id INTEGER PRIMARY KEY AUTOINCREMENT,  -- Unique identification number of the instructors
    lastname VARCHAR(255) NOT NULL,            -- Last name of the instructors
    firstname VARCHAR(255) NOT NULL,           -- First name of the instructors
    city VARCHAR(255),                         -- Name of the cities where instructors are located
    country CHAR(2)                            -- Two-letter country code of the countries where instructors are located
);
```

#### If no Where clause is used, all rows will be removeda

```python
%%sql
INSERT INTO Instructor(ins_id, lastname, firstname, city, country)
VALUES  (1,"Ahuja","Rav","Toronto","CA"),
        (2,"Chong","Raul","Toronto","CA"),
        (3,"Vasudevan","Hima","Chicago","US");
```

#### Insert a new instructor record with id 4 for Sandip Saha who lives in Edmonton, CA into the Instructor table.

```python
%%sql
INSERT INTO Instructor(ins_id, lastname, firstname, city, country)
VALUES(4, 'Saha', 'Sandip', 'Edmonton', 'CA');
```

```python
%%sql
SELECT * FROM Instructor;
```

### UPDATE

#### Update the city for Sandip to Toronto.



```python
%%sql
UPDATE Instructor 
SET city='Toronto' 
WHERE firstname="Sandip";

SELECT * FROM Instructor;
```

#### Update the city and country for Doe with id 5 to Dubai and AE respectively.

In this example, we want to update multiple columns of an existing row of the table.



```python
%%sql
UPDATE Instructor 
SET city='Dubai', country='AE' 
WHERE ins_id=5;

SELECT * FROM Instructor;
```

- The SQL retrieves specific data from databases.
- The COUNT, DISTINCT, and LIMIT are expressions used with SELECT statements. 
- The real-world applications of SELECT statements.
- The INSERT, UPDATE, and DELETE are DML statements for populating and changing tables
